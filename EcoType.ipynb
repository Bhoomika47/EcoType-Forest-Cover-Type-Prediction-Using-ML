```python
# Create copy of original dataframe
df_fe = df.copy()
```

```python
# Create total hillshade feature
df_fe["Total_Hillshade"] = (
    df_fe["Hillshade_9am"] +
    df_fe["Hillshade_Noon"] +
    df_fe["Hillshade_3pm"]
)
```

```python
# Create average hillshade feature
df_fe["Average_Hillshade"] = (
    df_fe["Total_Hillshade"] / 3
)
```

```python
# Create actual hydrology distance using Pythagoras formula
df_fe["Hydrology_Distance"] = np.sqrt(
    df_fe["Horizontal_Distance_To_Hydrology"]**2 +
    df_fe["Vertical_Distance_To_Hydrology"]**2
)
```

```python
# Combine road and fire point distances
df_fe["Road_Fire_Distance"] = (
    df_fe["Horizontal_Distance_To_Roadways"] +
    df_fe["Horizontal_Distance_To_Fire_Points"]
)
```

```python
# Create interaction feature
df_fe["Elevation_Slope"] = (
    df_fe["Elevation"] *
    df_fe["Slope"]
)
```

```python
# Create shade difference feature
df_fe["Shade_Difference"] = (
    df_fe["Hillshade_9am"] -
    df_fe["Hillshade_3pm"]
)
```

```python
# Display first 5 rows
df_fe.head()
```

```python
# Display updated dataframe shape
df_fe.shape
```

```python
# Check skewness in numeric columns
df_fe.skew(numeric_only=True)
```

```python
# Apply log transformation to skewed column
df_fe["Horizontal_Distance_To_Roadways"] = np.log1p(
    df_fe["Horizontal_Distance_To_Roadways"]
)
```

```python
# Store input features
X = df_fe.drop("Cover_Type", axis=1)

# Store target variable
y = df_fe["Cover_Type"]
```

```python
# Create Random Forest model
rf = RandomForestClassifier(
    random_state=42
)
```

```python
# Train model using dataset
rf.fit(X, y)
```

```python
# Import GridSearchCV
from sklearn.model_selection import GridSearchCV
```
